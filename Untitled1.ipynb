# 3. Perform Data Aggregation and Cleaning Required Join the dataframes together Identify and handle invalid values

```python
#merge the title basics dataframe with title ratings dataframe on common index 'tconst'
title_bsc_rts_df = pd.merge(title_bsc_df, title_rts_df, left_index=True, right_index=True)
title_bsc_rts_df
```

```python
#merge the title_bsc_rts dataframe with movie_gross dataframe on 'primary_title' and 'title'
boxoffice_df = pd.merge(title_bsc_rts_df, movies_gross_df, left_on ='primary_title', right_on = 'title')
boxoffice_df
```

```python
#remove duplicated columns
boxoffice_df.drop(columns = ['original_title', 'year'], axis = 1, inplace = True)
```

```python
boxoffice_df['genre']=boxoffice_df['genres'].str.split(",", expand= True)[0]
boxoffice_df['genre'][0]
```

```python
boxoffice_df['genre'].value_counts()
```

```python
#find out the distribution of movie genre

# Set up plots
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(16, 5))

# Create variables for easier reuse
value_counts = boxoffice_df['genre'].value_counts()
top_5_counts = value_counts.iloc[:5]

# Plot data
ax1.bar(value_counts.index, value_counts.values)
ax2.bar(top_5_counts.index, top_5_counts.values)

# Customize appearance
ax1.tick_params(axis="x", labelrotation=90)
ax2.tick_params(axis="x", labelrotation=45)
ax1.set_ylabel("Count of genres")
ax2.set_ylabel("Count of genres")
ax1.set_title("Distribution of films by genre")
ax2.set_title("Top 5 most produced genres  ");
```

```python
cols = ['start_year', 'runtime_minutes', 'averagerating', 'numvotes', 'domestic_gross', 'foreign_gross']
plt.figure(dpi=140)
sns.pairplot(boxoffice_df[cols])
plt.show()
```

```python
# Groupby multiple columns
result = boxoffice_df.groupby(['genre','averagerating','foreign_gross', 'domestic_gross'])['genre'].count()
print(result)
```

```python
# Groupby multiple columns & multiple aggregations
result = boxoffice_df.groupby('genre').aggregate({'averagerating':'sum','numvotes':'sum', 'domestic_gross':'sum', 'foreign_gross':'sum'})
print(result)
```

```python
import matplotlib.pyplot as plt

result.plot(kind="bar")
plt.xlabel("Genre")
plt.ylabel("sum Box Office")
plt.title("total Box Office by Genre")
plt.show()
```

```python
# Plotting line graph of the average rating
boxoffice_df.groupby('start_year')['averagerating'].sum().plot(kind='line')
plt.xlabel("Year")
plt.ylabel("Average Rating")
plt.title("Trend of Average Rating over Time")
plt.show()

# Plotting line graph of the number of votes
boxoffice_df.groupby('start_year')['numvotes'].sum().plot(kind='line')
plt.xlabel("Year")
plt.ylabel("Number of Votes")
plt.title("Trend of Number of Votes over Time")
plt.show()

# Plotting line graph of the domestic gross
boxoffice_df.groupby('start_year')['domestic_gross'].sum().plot(kind='line')
plt.xlabel("Year")
plt.ylabel("Domestic Gross")
plt.title("Trend of Domestic Gross over Time")
plt.show()

# Plotting line graph of the foreign gross
boxoffice_df.groupby('start_year')['foreign_gross'].sum().plot(kind='line')
plt.xlabel("Year")
plt.ylabel("Foreign Gross")
plt.title("Trend of Foreign Gross over Time")
plt.show()
```

```python
#Showing the distribution of the values for each of these columns by genre.

sns.boxplot(x='genre', y='averagerating', data=boxoffice_df)
plt.title('distribution of averagerating by genre')
plt.xticks(rotation=60)
plt.show()

sns.boxplot(x='genre', y='numvotes', data=boxoffice_df)
plt.title('distribution of numvotes by genre')
plt.xticks(rotation=60)
plt.show()

sns.boxplot(x='genre', y='domestic_gross', data=boxoffice_df)
plt.title('distribution of domestic_gross by genre')
plt.xticks(rotation=60)
plt.show()

sns.boxplot(x='genre', y='foreign_gross', data=boxoffice_df)
plt.title('distribution of foreign_gross by genre')
plt.xticks(rotation=60)
plt.show()
```
